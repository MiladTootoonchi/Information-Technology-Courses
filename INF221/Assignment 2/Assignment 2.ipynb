## Exercise 1  
The master method can be used to find the tight asymptotic bounds of divide-and-conquer algorithms. These runtimes take the form of $T(n) = aT(\frac{n}{b}) + f(n)$. We need to identify $a$ and $b$, then calculate the critical term $n^{\log_b{a}}$ and compare it to $f(n)$.  
  
The following subproblems have the same $a$ and $b$, meaning the critical term is $n^{\log_4{2}} = n^\frac{1}{2}$.
  
* a) $T(n) = 2T(\frac{n}{4}) + 1$  
$a = 2, b = 4, f(n) = 1$  
$f(n) = 1 = n^0 < n^\frac{1}{2} \Longrightarrow T(n) = \Theta(n^\frac{1}{2})$  
<br>
  
* b) $T(n) = 2T(\frac{n}{4}) + \sqrt{n}$  
$a = 2, b = 4, f(n) = \sqrt{n} = n^{\frac{1}{2}}$  
$f(n) = n^\frac{1}{2} = n^\frac{1}{2} \Longrightarrow T(n) = \Theta(n^{\log_4{2}} \log n) = \Theta(\sqrt{n} \log n)$  
<br>

* c) $T(n) = 2T(\frac{n}{4}) + n$  
$a = 2, b = 4, f(n) = n$  
$f(n) = n = n^1 > n^\frac{1}{2} \Longrightarrow T(n) = \Theta(n)$  
<br>

* d) $T(n) = 2T(\frac{n}{4}) + n^2$  
$a = 2 , b = 4, f(n) = n^2$  
$f(n) = n^2 > n^\frac{1}{2} \Longrightarrow T(n) = \Theta(n^2)$

````python
import random

def permute_without_identity(A):
    n = len(A)
    A = A.copy()  # making sure the original list isn't changing
    
    for i in range(n - 1):               # i = 0 to n-2 (Python indexing)
        j = random.randint(i + 1, n - 1) # Random(i+1, n)
        A[i], A[j] = A[j], A[i]          # swap
    
    return A
````

Since the random index is always chosen from i+1 to n, an element is never swapped with itself. Therefore, the original ordering of A can never be returned.

P([3, 2, 1]) = 1/2

Case 2:
1. Swap A[1] with A[3] and you get [3, 2, 1]

P([3, 1, 2]) = 1/2

<br>

The algorithm can only make the following permutations:
1. [2, 3, 1]
2. [3, 1, 2]

However, for three elements there are 6 total permutations since 3! = 6.
If we exclude the original ordering [1, 2, 3], there are 5 possible permutations.

Since the algorithm only generates 2 of the 5 possible permutations, several permutations can never occur.
The algorithm guarantees that the original ordering is never returned. However, it does not generate all possible permutations of A (except the original) with equal probability. Therefore, the algorithm is not correct.

````python
import heapq

def mergeLists(lists):
    '''
    Merges k sorted lists into one sorted list using a min-heap.
    
    input: lists - a list of k sorted lists (total n elements)
    
    output: a single sorted list containing all n elements from the input lists
    '''
    
    min_heap = []
    result = []

    # Get numbers form lists to heap
    for idx, list in enumerate(lists):
        if list:
            heapq.heappush(min_heap, (list[0], idx, 0))

    # pop push cycle into result
    while min_heap: # <- end when heap empty
        value, list_idx, elem_idx = heapq.heappop(min_heap)
        result.append(value)

        # push next element from same list
        if elem_idx + 1 < len(lists[list_idx]):
            nextValue = lists[list_idx][elem_idx + 1]
            heapq.heappush(min_heap, (nextValue, list_idx, elem_idx + 1))

    return result
````

````python
lists = [
    [1, 5, 7],
    [2, 4],
    [3, 6, 8, 9]
]

print(mergeLists(lists))
````

### 1. 
If Quicksort always chooses the median as the pivot, the array is split into two equal halves at every step. The recurrence relation becomes

T(n) = 2T(n/2) + theta(n)

The theta(n) term comes from the partition operation.

Using the Master Theorem:

* a = 2
* b = 2
* f(n) = n

n^(log_b(a)) = n^(log_2(2)) = n

Since f(n) = theta(n^(log_b(a))), this is case 2 of the Master Theorem and therefore:

T(n) = theta(n * log(n))

So if quicksort always partitions on the median, its runtime is theta(n * log(n)) in all cases.

### 2.
To always partition on the median, the algorithm must first find the median of the current subarray. Finding the median by sorting would cost O(n * log(n)), which would make the algorithm not efficient, but the Median-of-Medians algorithm can find the median in O(n) time. If we use this algorithm, each recursive step performs:

* Median selection: O(n)

* Partitioning: O(n)

Thus the recurrence becomes:

* T(n) = 2T(n/2) + O(n)

which still solves to:

* T(n) = theta(n * log(n))

Therefore, it can be made efficiently, although in practice this approach is rarely used because the median-of-medians algorithm has a large constant factors, making randomized quicksort faster in practice.
